```python
#Checking if dataset loaded properly
import pandas as pd
df = pd.read_csv('netflix_titles.csv')

df.head()
```

```python
#Converting column to 'datetime' format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
#Create a new column for 'year'
df['year_added'] = df['date_added'].dt.year
df.head()
```

```python
#Counting how many times same year shows up
df['year_added'].value_counts()
```

```python
#Bar graph for an easy demonstration
df['year_added'].value_counts().plot(kind='bar')
```
