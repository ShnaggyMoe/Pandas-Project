```python
#Loading the datasets
import pandas as pd

df = pd.read_csv('nba_2022-23_all_stats_with_salary.csv')
df.head()

nba_df = df[['Player Name' , 'PTS' , 'Salary']].copy()

import plotly.express as px
px.scatter()

chart = px.scatter(nba_df, x ='Salary', y ='PTS', hover_data=['Player Name'])
chart.show()
```
